```python
import pandas as pd

dataset = pd.read_csv("dataset8.csv", index_col="index")
print(dataset)
dataset.apply(pd.isnull).sum() / dataset.shape[0]
dataset['date_time'] = pd.to_datetime(dataset['date_time'], format='%d-%m-%Y %H:%M')
dataset['time'] = dataset['date_time'].dt.hour
dataset['date_time'] = dataset['date_time'].dt.month
dataset.columns = dataset.columns.str.strip()
mainDataset = dataset[
    ['date_time', 'time', 'latitude', 'longitude', 'tempC', 'FeelsLikeC', 'mintempC', 'maxtempC', 'pressure',
     'humidity', 'visibility', 'windspeedKmph', 'winddirDegree', 'WindGustKmph', 'precipMM', 'cloudcover']].copy()
mainDataset.head()
import matplotlib.pyplot as plt

plt.plot(mainDataset['date_time'], mainDataset['latitude'], label='Latitude')
plt.plot(mainDataset['date_time'], mainDataset['longitude'], label='Longitude')
plt.plot(mainDataset['date_time'], mainDataset['tempC'], label='Temperature')
plt.legend()
plt.show()
```
